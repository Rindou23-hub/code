```python
# @title Program Binary Tree Inorder
# Struktur Tree berdasarkan identitas saya
class Node:
    def __init__(self, data, label=""):
        self.data = data
        self.label = label
        self.left = None
        self.right = None

def inorder_traversal(node):
    if node:
        inorder_traversal(node.left)
        print(f"{node.label}: {node.data}")
        inorder_traversal(node.right)

# Data identitas saya
nama = "Kartika Salsabila"
npm = "2410506003"
tanggal_lahir = "12-08-2006"

jumlah_huruf = len(nama.replace(" ", ""))

vokal = "aiueoAIUEO"
jumlah_vokal = sum(1 for huruf in nama if huruf in vokal)

dua_digit_npm = int(npm[-2:])

tanggal = int(tanggal_lahir.split("-")[0])
bulan = int(tanggal_lahir.split("-")[1])

root = Node(jumlah_huruf, "+ Root \n(Jumlah huruf nama)")
root.left = Node(jumlah_vokal, "+ Anak Kiri Root \n(Jumlah huruf vokal)")
root.right = Node(dua_digit_npm, "+ Anak Kanan Root \n(2 digit NPM)")

root.left.left = Node(tanggal, "+ Anak Kiri dari Sebelah Kiri \n(DD)")
root.right.right = Node(bulan, "+ Anak Kanan dari Sebelah Kanan \n(MM)")

# Menampilkan tree dengan inorder traversal
print("Inorder Traversal dari Tree:\n")
inorder_traversal(root)
```

```python
# @title Program BST
# Program BST dari data numerik unik berdasarkan identitas saya

#class Node untuk struktur BST
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

#menyisipkan data ke dalam BST
def insert(root, data):
    if root is None:
        return Node(data)
    if data < root.data:
        root.left = insert(root.left, data)
    else:
        root.right = insert(root.right, data)
    return root

#traversal BST
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end=' ')
        inorder(root.right)

def preorder(root):
    if root:
        print(root.data, end=' ')
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data, end=' ')

#searching nilai dalam BST
def search(root, key):
    if root is None:
        return False
    if root.data == key:
        return True
    elif key < root.data:
        return search(root.left, key)
    else:
        return search(root.right, key)

# === DATA IDENTITAS ===
npm = "2410506003"
nama = "Kartika Salsabila"
tanggal_lahir = "12-08-2006"

data_bst = []

# 1. Tiga digit terakhir NPM
data_bst.append(int(npm[-3:]))

# 2. Tanggal lahir: DD dan MM
tanggal = int(tanggal_lahir.split("-")[0])
bulan = int(tanggal_lahir.split("-")[1])
data_bst.extend([tanggal, bulan])

# 3. ASCII dari huruf pertama dan kedua nama depan
ascii_1 = ord(nama[0])
ascii_2 = ord(nama[1])
data_bst.extend([ascii_1, ascii_2])

# 4. Dua angka bebas
data_bst.extend([15, 25])

# Menampilkan data
print("Data numerik unik:", data_bst)

# === MEMBANGUN BST ===
root = None
for num in data_bst:
    root = insert(root, num)

# === MENAMPILKAN TRAVERSAL ===
print("\nInorder Traversal: ")
inorder(root)

print("\n\nPreorder Traversal: ")
preorder(root)

print("\n\nPostorder Traversal: ")
postorder(root)

# === FITUR PENCARIAN ===
print("\n\n=== Fitur Pencarian ===")
target1 = int(npm[-2:])  # dua digit terakhir NPM: 03
target2 = 100           # angka yang dipastikan tidak ada

print(f"\nMencari angka {target1}...")
print("Ditemukan!" if search(root, target1) else "Tidak ditemukan.")

print(f"\nMencari angka {target2}...")
print("Ditemukan!" if search(root, target2) else "Tidak ditemukan.")
```
