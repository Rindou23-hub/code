```python
import pandas as pd
import numpy as np

# Contoh dataset dengan beberapa masalah data
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'tablet', 'Phone', None],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets', 'Electronics', 'Gadgets'],
    'Price': [1200, 1300, None, 800, 800, 600, 600],
    'Quantity': [5, None, 8, 3, 3, 10, None]
}

df = pd.DataFrame(data)
print("Dataset Awal")
print(df)

print("\nMissing Values")
print(df.isnull().sum())

df['Price'].fillna(df['Price'].mean(), inplace=True)       
df['Quantity'].fillna(df['Quantity'].median(), inplace=True) 

df.dropna(subset=['Product'], inplace=True)

print("\nMenangani Missing Data")
print(df)

num_cols = ['Price', 'Quantity']
df[num_cols] = (df[num_cols] - df[num_cols].min()) / (df[num_cols].max() - df[num_cols].min())

print("\nNormalisasi Numerik")
print(df)

df['Product'] = df['Product'].str.strip().str.capitalize()
df['Category'] = df['Category'].str.strip().str.capitalize()

print("\nStandarisasi Kategorikal")
print(df)

df.drop_duplicates(inplace=True)
print("\nMenghapus Duplikasi")
print(df)
```
