```python
# Kelas Buku
class Buku:
    def __init__(self, judul, pengarang, tahun_terbit, jumlah_stok):
        self.judul = judul
        self.pengarang = pengarang
        self.tahun_terbit = tahun_terbit
        self.jumlah_stok = jumlah_stok

    def pinjam(self):
        if self.jumlah_stok > 0:
            self.jumlah_stok -= 1
            print(f"Buku '{self.judul}' berhasil dipinjam. Stok tersisa: {self.jumlah_stok}")
        else:
            raise ValueError("Stok buku habis. Tidak dapat dipinjam.")

    def kembalikan(self):
        self.jumlah_stok += 1
        print(f"Buku '{self.judul}' berhasil dikembalikan. Stok sekarang: {self.jumlah_stok}")


# Fungsi untuk menampilkan semua buku
def tampilkan_semua_buku(daftar_buku):
    if daftar_buku:
        print("\nDaftar Buku di Perpustakaan:")
        for buku in daftar_buku:
            print(f"Judul: {buku.judul}, Pengarang: {buku.pengarang}, Tahun Terbit: {buku.tahun_terbit}, Stok: {buku.jumlah_stok}")
    else:
        print("Tidak ada buku di perpustakaan.")


# Fungsi untuk mencari buku berdasarkan judul
def cari_buku(daftar_buku, judul):
    for buku in daftar_buku:
        if buku.judul.lower() == judul.lower():
            return buku
    raise ValueError(f"Buku dengan judul '{judul}' tidak ditemukan.")


# Fungsi untuk menambah buku baru
def tambah_buku(daftar_buku, id_buku, buku):
    daftar_buku.append(buku)
    id_buku[buku.judul] = buku
    print(f"Buku '{buku.judul}' telah ditambahkan.")


# Fungsi untuk meminta input jumlah stok dengan validasi
def input_jumlah_stok():
    while True:
        try:
            jumlah_stok = int(input("Masukkan jumlah stok: "))  # Cek apakah input berupa angka
            if jumlah_stok < 0:
                raise ValueError("Jumlah stok tidak boleh negatif.")
            return jumlah_stok
        except ValueError:
            print("Input harus berupa angka aja. Coba lagi.")  # Pesan kesalahan yang lebih sederhana


# Fungsi untuk meminta input tahun terbit dengan validasi
def input_tahun_terbit():
    while True:
        try:
            tahun_terbit = int(input("Masukkan tahun terbit buku: "))  # Cek apakah input berupa angka
            if tahun_terbit < 0:
                raise ValueError("Tahun terbit tidak boleh negatif.")
            return tahun_terbit
        except ValueError:
            print("Input harus berupa angka aja. Coba lagi.")  # Pesan kesalahan yang lebih sederhana


# Fungsi utama untuk menjalankan menu
def menu():
    daftar_buku = []
    id_buku = {}

    while True:
        print("\n=== Menu Perpustakaan ===")
        print("1. Tambah Buku")
        print("2. Pinjam Buku")
        print("3. Kembalikan Buku")
        print("4. Cari Buku")
        print("5. Tampilkan Semua Buku")
        print("6. Keluar")

        pilihan = input("Pilih menu: ")

        try:
            if pilihan == "1":
                # Tambah buku baru
                judul = input("Masukkan judul buku: ")
                pengarang = input("Masukkan pengarang buku: ")
                tahun_terbit = input_tahun_terbit()  # Menggunakan fungsi untuk input tahun terbit dengan validasi
                jumlah_stok = input_jumlah_stok()  # Menggunakan fungsi untuk input jumlah stok dengan validasi
                buku_baru = Buku(judul, pengarang, tahun_terbit, jumlah_stok)
                tambah_buku(daftar_buku, id_buku, buku_baru)

            elif pilihan == "2":
                # Pinjam buku
                judul = input("Masukkan judul buku yang ingin dipinjam: ")
                buku = cari_buku(daftar_buku, judul)
                buku.pinjam()

            elif pilihan == "3":
                # Kembalikan buku
                judul = input("Masukkan judul buku yang ingin dikembalikan: ")
                buku = cari_buku(daftar_buku, judul)
                buku.kembalikan()

            elif pilihan == "4":
                # Cari buku
                judul = input("Masukkan judul buku yang ingin dicari: ")
                buku = cari_buku(daftar_buku, judul)
                print(f"Judul: {buku.judul}, Pengarang: {buku.pengarang}, Tahun Terbit: {buku.tahun_terbit}, Stok: {buku.jumlah_stok}")

            elif pilihan == "5":
                # Tampilkan semua buku
                tampilkan_semua_buku(daftar_buku)

            elif pilihan == "6":
                print("Terima kasih telah menggunakan program manajemen perpustakaan!")
                break

            else:
                raise ValueError("Pilihan tidak valid.")

        except ValueError as e:
            print(f"Mohon maaf {e}")
        except Exception as e:
            print(f"Terjadi kesalahan yang tidak terduga: {e}")


# Menjalankan menu utama
menu()
```
